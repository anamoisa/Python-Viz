```
# Your code here
df_weather['DATE']=pd.to_datetime(df_weather['DATE'])
fig, (ax1, ax2) = plt.subplots(1, 2,figsize=(10,5))
fig.suptitle('Summer versus Winter')

df_season1=df_weather.loc[(df_weather['MONTH']>6) &(df_weather['MONTH']<10)]
df_season2=df_weather.loc[(df_weather['MONTH']>=0) &(df_weather['MONTH']<5)]

sns.lineplot(data=df_season1, x='DATE', y='MAX_TEMPERATURE_C',ci=None, 
             color='rebeccapurple', label='Summer', ax=ax1)
sns.lineplot(data=df_season2,x='DATE', y='MAX_TEMPERATURE_C', ci=None, 
             color='mediumorchid', label='Winter', ax=ax1)

sns.scatterplot(x=df_season1['DATE'], y=df_season1['MAX_TEMPERATURE_C'],
            color='crimson', label='Summer', ax=ax2)
sns.scatterplot(x=df_season2['DATE'], y=df_season2['MAX_TEMPERATURE_C'],
           color='darkmagenta', label='Winter', ax=ax2)

ax2.xaxis.set_major_locator(mdates.MonthLocator(interval=1))
ax2.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
plt.gcf().autofmt_xdate()
#ax.XAxis.set_ticks(pd.date_range(start='2019-01-01', freq='MS', periods=12))
#ax.XAsis.set_major_formatter(mdates.DateFormatter('%Y_%m'))
plt.show()

```
